###Double linked list

 - 이중 연결 리스트
 - single linked list의 가장 큰 단점인 앞쪽 노드를 찾기 어렵다는 점을 개선
 - data/prev/next 로 구성
 - 이 아이는 알고리즘 문제에도 꽤 나옴

리스트로만 풀면 current row 업데이트가 제대로 안됨

```python
def solution(n, k, cmd):
    answer = [i for i in range(n)]
    z = []
    for c in cmd:
        idx = c.split()
        if idx[0] == 'D':
            k = k+int(idx[1])
            if k not in answer:
                while k in answer:
                        k += 1
            print(k)
                
        elif idx[0] == 'U':
            k = k- int(idx[1])
            if k not in answer:
                while k in answer:
                    k -= 1
            print(k)           
        elif idx[0] == 'C':
            z.append(answer.pop(k))
            if k != n-1:
                k -= 1
            else:
                k += 1
        else:
            num = z.pop(-1)
            answer.append(num)
            answer.sort()
        
    return ''.join(map(lambda x:'O' if x in answer else 'X',range(0,n)))
```

###Circular linked list

image.png

꼬리 노드가 다시 머리 노드를 가르키는 구조

```python
class Node:
  def __init__(self, element):
    self.element = element
    self.next = None

class CircularLinkedList:
  def__init__(self):
    self.head = Node('head')
    self.head.next = self.head
```
